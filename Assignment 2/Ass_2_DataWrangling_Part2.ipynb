## Finding total count and indexes of missing values in attendance fields and replacing them with mean value.

```python
print('Total Rows with missing Current Back : ',format(df['Current Back'].head(100).isna().sum()))
indices_attendance = []
for index, row in df.iterrows():
     if(pd.isnull(row['Current Back'])):
         indices_attendance.append(index)  
print('Indices : ',indices_attendance)
```

```python
att_mean = df['Current Back'].mean()
df['Current Back'].fillna(att_mean, inplace= True)
df['Current Back'] = df['Current Back'].round(1)
```

## Scaling CGPA values from the scale of 100 to the scale of 0 -> 10 using MinMaxScaler(min,max)

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,10))
df[['GPA 1']] = scaler.fit_transform(df[['GPA 1']])
df['GPA 1'] = df['GPA 1'].round(1)
```
